```python
'''This method compares the proportions of one categorical variable with the categories of another variable'''

'''Args
rotate: rotate the xtick labels
flip: swap the variable on the x-axis with the proportion variable
rs: remove sparse data (where total < specified value)
percent: shows percentage of columns, set to False to change to histogram-based counts
'''

def proportion(x, prop, df, flip = False, rotate=0, rs=0, percent=True):  
  fs = (6,12)

  # transforms df into idx: (0, 1) and cols: prop
  r = df.groupby([prop, x]).count().unstack().fillna(0)['id']
  # orders by fraction
  r['frac'] = r[0]/r[1]
  r = r.sort_values(by = 'frac')
  r = r.drop('frac', 1)


  if rs:
    r = r[r[0]+r[1] > rs]
  if flip:
    fs = (20,6)
    r = r.T
  r = r.T

  # sums each row in the r dataframe
  totals = list(r.sum(axis=1).values)
  cols = list(r.columns)

  # list to store proportions
  pl = []
  for col in cols:
      pl.append([i / j * 100 for i,j in zip(r[col], totals)])

  fig, ax = plt.subplots(figsize=fs)
  text_size = 12
  
  if percent:
    for i in range(len(pl)):
      bot = [sum(i) for i in zip(*pl[:i])]
      if not bot:
          bot = 0
      ax.bar(r.index, pl[i], bottom = bot, label = cols[i])
    ax.set_ylabel('percentage', size=text_size)
    ax.set_title(str(prop)+' proportion based on '+str(x), size=text_size+2)
  else:
    for i in range(len(cols)):
      if i==0:
        ax.bar(r.index, r[cols[i]], label = cols[i])
      else:
        ax.bar(r.index, r[cols[i]], bottom = r[cols[0:i]].sum(axis=1), label = cols[i])
    ax.set_ylabel('count', size=text_size)
    ax.set_title('proportion of '+str(prop)+' counts based on '+str(x), size=text_size+2)

  ax.set_xlabel(prop, size=text_size)

  handles, labels = ax.get_legend_handles_labels()
  ax.legend(reversed(handles), reversed(labels), bbox_to_anchor=(1.02, 1), title=x)
  plt.xticks(rotation=rotate)

  plt.show()
```

I removed the quantity_group column because it's identical to quantity. And I remoeved payment_type because it is grouped identically to payment (ex: 'pay_monthly' in payment always matches with 'monthly' in payment type). 'date_recorded' was dropped since I already extracted data from it.

```python
train['month'] = [calendar.month_abbr[i.month] for i in dates]
train['day'] = ['d'+str(i.day) for i in dates]
print(len(train)-sum(sum([train['quantity']==train['quantity_group']])))
train1 = train.drop(columns=['quantity_group', 'payment_type', 'date_recorded', 'id'])
train1[0:6]
```

Splitting data into training (60%), validation (20%), and testing (20%) sets. I will use target encoding due to the high dimensionality of the data, then scale it. I will also add a column of ones (this will be for the bias) which will be used in the LogisticRegression model below.

```python

y = train1['status_group']
y = np.array([y]).T # transforming the dimensions of y into a vertical vector (m x 1)

train2 = train1.drop(columns=['status_group'])

cols = []
# find categorical collumns
for col in train1.columns:
  for elem in train1[col]:
    if elem!=elem:
      pass
    elif type(elem)==str:
      cols.append(col)
      break
    else:
      break



X_dev, X_test, y_dev, y_test = train_test_split(train2, y, test_size=0.2, random_state=0)
X_train, X_val, y_train, y_val = train_test_split(X_dev, y_dev, test_size=0.25, random_state=0)


# encode the categorical columns
te = TargetEncoder()
X_train[cols] = te.fit_transform(X_train[cols], y_train)
X_val[cols] = te.transform(X_val[cols])
X_test[cols] = te.transform(X_test[cols])


# impute missing values
imp =  SimpleImputer(strategy='median')
X_train = imp.fit_transform(X_train)
X_val = imp.transform(X_val)
X_test = imp.transform(X_test)


# scale data on training, then apply scaling on validation and testing data
ss = StandardScaler()
X_train = ss.fit_transform(X_train)   # Fit and transform scalar on X_train
X_val = ss.transform(X_val)           # Transform X_val
X_test = ss.transform(X_test)         # Transform X_test


# append column of ones to the features
X_train = np.hstack([np.ones((X_train.shape[0], 1)), X_train])
X_val = np.hstack([np.ones((X_val.shape[0], 1)), X_val])
X_test = np.hstack([np.ones((X_test.shape[0], 1)), X_test])
```

### Model Building

```python
class LogisticRegression():
  '''
  Logistic regression model with L2 regularization

  Attributes
  ----------
  alpha: parameter for regularization
  t: # epochs to run gradient descent
  eta: learning rate for gradient descent

  Vectors
  ----------
  X: (m x n) feature matrix
  w: (n x 1) weight vector
  y: (m x 1) label vector
  '''
  def __init__(self, alpha=0, t=100, eta=1e-3):
    self.alpha = alpha
    self.t = t
    self.eta = eta


  def train(self, X, y):
    '''
    Trains model with gradient descent by setting w
    '''
    self.w = np.array([X.shape[1]*[0]]).T    
    # start gradient descent to find w
    for i in range(self.t):
      self.w = self.w - (self.eta * self.gradient_var(X,y))
    return


  def gradient_var(self, X, y):
    '''
    Computes and returns gradient vector (n x 1) given X (m x n) and y (m x 1)
    '''
    return (X.T@(self.sigmoid(X@self.w) - y)) + (2 * self.alpha * self.w)


  def sigmoid(self, x):
    '''
    returns sigmoid on every element in the x (m x 1) vector
    '''
    return 1/(1+np.exp(-x))

  
  def predict(self, X):
    '''
    Return a probability prediction vector (m x 1) given X (m x n)
    '''
    return self.sigmoid(X@self.w)
```

### Model Evaluation

Grid search to find optimal regularization parameter ($\alpha$) and learning rate ($\eta$).

```python
# Add code here to evalute your models.
min_log = 99999
min_alpha = None
min_eta = None

alphas = np.logspace(-2,3,50)
etas = np.logspace(-7,-5,50)
for a in alphas:
  for e in etas:
    # split development data into 
    lr = LogisticRegression(alpha=a, eta=e)
    lr.train(X_train, y_train)
    pred = lr.predict(X_val)
    ll = log_loss(y_val, pred)
    if ll < min_log:
      min_log = ll
      min_alpha = a
      min_eta = e

print('min log:', min_log, '  best alpha:', min_alpha, '    best eta:', min_eta)
```

```python
linear_reg = LogisticRegression(alpha = min_alpha, eta = min_eta)
linear_reg.train(X_train, y_train)
print('Log loss on testing dataset:')
log_loss(y_test, linear_reg.predict(X_test))
```
